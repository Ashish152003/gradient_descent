```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,random_state=13,noise=80)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x))
```

```python
y_pred = (78.35063668*x + 0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='ols')
plt.plot(x,y_pred,color='blue',label='b')
plt.legend()
plt.show()
```

```python
m=78.35063668
b=-100
lr=0.01
eposch=100
for i in range(eposch):
    loss_slope = -2*np.sum(y-m*x.ravel()-b)
    step_size=lr*loss_slope
    b=b-step_size
    y_pred = m*x + b
    plt.scatter(x,y)
    plt.plot(x,y_pred)
```
